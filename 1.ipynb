```python
def calculate_sma(df, window_size):
    """
    Calculates the Simple Moving Average for a given window.
    :param df: Pandas DataFrame containing a 'Close' column.
    :param window_size: Integer (e.g., 20, 50) for the average period.
    :return: DataFrame with a new SMA column.
    """
    column_name = f'SMA_{window_size}'
    
    # .rolling(window) looks back at the specified number of rows
    # .mean() calculates the average of those rows
    df[column_name] = df['Close'].rolling(window=window_size).mean()
    
    return df

# Example Usage:
# df = pd.read_csv('nabil_data.csv')
df = calculate_sma(data, 1)  # Adds 'SMA_20' column
print("hey man sma is :")
print(df)
# df = calculate_sma(df, 50)  # Adds 'SMA_50' column
```
